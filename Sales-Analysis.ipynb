```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load The Data

```python
# Load The Data

data = pd.read_csv('C:/Users/DELL/OneDrive/Brainwave_Intern/Superstore.csv',encoding='ISO-8859-1')
pd.set_option('display.max_columns', None)
data
```

# Data Exploration

```python
print(data.head())
print(data.info())
data.describe()
```

```python
# check missing values

print(data.isnull().sum())
```

# Data Cleaning

```python
data.drop(columns=['Row ID', 'Postal Code'], inplace=True)
data.dropna(inplace=True)
```

```python
data.head()
```

```python
# convert date to datetime

data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data.head()
```

# Data Analysis

## Sales Overview

```python
# Total Sales

total_sales = data['Sales'].sum()
print(f"Total Sales: ${total_sales:,.2f}")
```

```python
# Total Profit

total_profit = data['Profit'].sum()
print(f"Total Profit: ${total_profit:,.2f}")
```

```python
# Total Quantity

total_quantity = data['Quantity'].sum()
print(f"Total Quantity Sold: {total_quantity}")
```

## Top 5 Profitable Products

```python
top_products = data.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(5)
print(top_products)
```

## Category-wise Analysis

```python
# sales by category

category_sales = data.groupby('Category')['Sales'].sum()

plt.figure(figsize=(10,5))
sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis')
plt.title('Sales by Category')
plt.show()
```

```python
# profit by category

category_profit = data.groupby('Category')['Profit'].sum()

plt.figure(figsize=(10,5))
sns.barplot(x=category_profit.index, y=category_profit.values, palette='viridis')
plt.title('Profit by Category')
plt.show()
```

## Region-wise Performance

```python
# Sales by region
region_sales = data.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=region_sales.index, y=region_sales.values, palette='coolwarm')
plt.title('Sales by Region')
plt.show()
```

```python
# Profit by region

region_profit = data.groupby('Region')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=region_profit.index, y=region_profit.values, palette='coolwarm')
plt.title('Profit by Region')
plt.show()
```

## Monthly Sales Trend

```python
data['Month'] = data['Order Date'].dt.to_period('M')
monthly_sales = data.groupby('Month')['Sales'].sum()

plt.figure(figsize=(15,8))
monthly_sales.plot(marker='o', color='orange')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales ($)')
plt.grid()
plt.show()
```
